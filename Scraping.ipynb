```python
import requests
from bs4 import BeautifulSoup
import os
from selenium import webdriver
import time


def get_thumbnails(search_term, count=100, download_folder="photos"):
    thumbnails = []

    try:
        url = rf'https://www.google.no/search?q={search_term}&client=opera&hs=cTQ&source=lnms&tbm=isch&sa=X&safe=active&ved=0ahUKEwig3LOx4PzKAhWGFywKHZyZAAgQ_AUIBygB&biw=1920&bih=982'

        # Set headers to mimic a browser request
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

        page = requests.get(url, headers=headers).text
        soup = BeautifulSoup(page, 'html.parser')

        # Create folder to save thumbnails if it doesn't exist
        if not os.path.exists(download_folder):
            os.makedirs(download_folder)

        # Initialize Chrome WebDriver
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")  # To run Chrome in headless mode (without opening GUI)
        driver = webdriver.Chrome(options=options)

        # Use the WebDriver to open the URL
        driver.get(url)

        # Simulate scrolling down the page to load more images
        for _ in range(10):  # Adjust the range according to your needs
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            for i, raw_img in enumerate(soup.find_all('img')):
                link = raw_img.get('src')
                if link and link.startswith("https://"):
                    thumbnails.append(link)
                    # Get the full-size image URL
                    full_size_link = link.replace('w160-h160', 'w0-h0')
                    # Download the image
                    img_data = requests.get(full_size_link, headers=headers).content
                    with open(os.path.join(download_folder, f"{search_term}_{i}.jpg"), 'wb') as handler:
                        handler.write(img_data)

                    if len(thumbnails) >= count:
                        driver.quit()  # Close the WebDriver after gathering required thumbnails
                        return thumbnails
            # Scroll down to load more images
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Adjust the sleep time as needed

        driver.quit()  # Close the WebDriver after gathering thumbnails if the loop ends
        return thumbnails
    except Exception as e:
        sError = f"""Error: {e}
        search_term: {search_term}
        """
        print(sError)
        return thumbnails
```

```python
thumbnails = get_thumbnails("Ayam Goreng", count=1000, download_folder="photos/Ayam Goreng")
for i, thumbnail in enumerate(thumbnails):
    print(thumbnail)
```
